```python
import pandas as pd

data = {
    "이름" : ["이지은", "박동훈", "홍길동", "강감찬", "오해영"],
    "학과" : ["컴퓨터", "기계", "철학", "컴퓨터", "철학"],
    "학년" : [1, 2, 2, 4, 3],
    "학점" : [1.5, 2.0, 3.1, 1.1, 2.7]
}

df = pd.DataFrame(data,columns=["이름","학과","학년","학점","등급"],
                 index = ["one","two","three","four","five"])

display(df)

# 1. 이름이 박동훈인 사람을 찾아 이름과 학점을 DataFrame으로 출력
P1 = df.loc[df["이름"]=="박동훈",["이름","학점"]]
display(P1)

# 2. 학점이 (1.5, 2.5)인 사람을 찾아 학과, 이름, 학점을 DataFrame으로 출력
P2 = df.loc[(df["학점"]>1.5) & (df["학점"]<2.5),["학과","이름","학점"]]
display(P2)

# 3. 학점이 3.0을 초과하는 사람을 찾아 등급을 'A'로 설정
df.loc[df["학점"]>3.0, "등급"] = 'A'
display(df)
```

# Pandas Exercis - 1

```python
# pandas를 사용하여 다음의 문제를 해결합니다. 동률이 있을 경우 오름차순으로 정렬하여 출력합니다.

import pandas as pd
import numpy as np

links = pd.read_csv("ml-latest-small/links.csv")
movies = pd.read_csv("ml-latest-small/movies.csv")
ratings = pd.read_csv("ml-latest-small/ratings.csv")
tags = pd.read_csv("ml-latest-small/tags.csv")

# 1. 사용자가 평가한 모든 영화의 전체 평균 평점
ratingAvg = ratings['rating'].sum()/len(ratings)
print("사용자가 평가한 모든 영화의 전체 평균 평점 : {:.5f}".format(ratingAvg) )
```

```python
# 2. 각 사용자별 평균 평점
userRatings = ratings['rating'].groupby(ratings['userId'])
print(userRatings.mean())
```

```python
# 3. 각 영화별 평균 평점
movieRatings = ratings['rating'].groupby(ratings['movieId'])
df = pd.merge(movieRatings.mean(),movies.loc[:,'movieId':'title'],on='movieId',how='right')
display(df)
```

```python
# 4. 평균 평점이 가장 높은 영화의 제목(동률이 있을 경우 모두 출력)
df = df.loc[df['rating'] == df['rating'].max(), :]
df.sort_values(by='title',inplace=True)
df
```

```python
# 5. Comedy영화 중 가장 평점이 낮은 영화의 제목
df2 = pd.merge(movieRatings.mean(),movies,on='movieId',how='inner')
comedyMovies = df2.loc[df2['genres'].str.contains('Comedy'),:]
comedyMovies = comedyMovies.loc[comedyMovies['rating'] == comedyMovies['rating'].min(), :]
comedyMovies.sort_values(by='title',inplace=True)
comedyMovies
```

```python
# 6. 2015년도에 평가된 모든 Romance 영화의 평균 평점은?
import time
from datetime import datetime

start = time.mktime(datetime.strptime('2015-01-01 00:00:00', '%Y-%m-%d %H:%M:%S').timetuple())
end = time.mktime(datetime.strptime('2016-01-01 00:00:00', '%Y-%m-%d %H:%M:%S').timetuple())

rating2015 = ratings.loc[(ratings['timestamp'] >= start) & (ratings['timestamp'] < end), :]
rating2015 = pd.merge(rating2015,movies,on='movieId',how='inner')

romance2015 = rating2015.loc[rating2015['genres'].str.contains('Romance'),:]

ratingAvg = romance2015['rating'].sum()/len(romance2015)
print("2015년도에 평가된 모든 Romance 영화의 평균 평점 : {:.5f}".format(ratingAvg) )
```

# Pandas Exercis - 2 

```python
import pandas as pd
import numpy as np

mpg = pd.read_csv('mpg.txt')

# 1. displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중 
#    어떤 자동차의 hwy(고속도로 연비)가 평균적으로 더 높은지 확인하세요.

displ4 = mpg.loc[mpg['displ'] <= 4, :]
displ5 = mpg.loc[mpg['displ'] >= 5, :]

print("배기량 4이하인 자동차의 고속도로 평균 연비 : {:.5f}".format(displ4['hwy'].mean()))
print("배기량 5이상인 자동차의 고속도로 평균 연비 : {:.5f}".format(displ5['hwy'].mean()))
```

```python
# 2. 자동차 제조 회사에 따라 도시 연비가 다른지 알아보려고 한다. 
#    "audi"와 "toyota" 중 어느 manufacturer(제조회사)의 cty(도시 연비)가 
#    평균적으로 더 높은지 확인하세요.

audi = mpg.loc[mpg['manufacturer'] == 'audi', :]
toyota = mpg.loc[mpg['manufacturer'] == 'toyota', :]

print("Audi의 도시 평균 연비 : {:.5f}".format(audi['cty'].mean()))
print("Toyota의 도시 평균 연비 : {:.5f}".format(toyota['cty'].mean()))
```

```python
# 3. "chevrolet", "ford", "honda" 자동차의 고속도로 연비 평균을 알아보려고 한다. 
#    이 회사들의 데이터를 추출한 후 hwy(고속도로 연비) 평균을 구하세요.

cfh = mpg.loc[(mpg['manufacturer'] == 'chevrolet') |
              (mpg['manufacturer'] == 'ford') |
              (mpg['manufacturer'] == 'honda'), :]

print("Chevrolet, Ford, Honda의 도시 평균 연비 : {:.5f}".format(cfh['hwy'].mean()))
```

```python
# 4. "audi"에서 생산한 자동차 중에 어떤 자동차 모델의 hwy(고속도로 연비)가 
#    높은지 알아보려고 한다. 
#    "audi"에서 생산한 자동차 중 hwy가 1~5위에 해당하는 자동차의 데이터를 출력하세요.

audi = mpg.loc[mpg['manufacturer'] == 'audi', :]
audi = audi.sort_values(by='hwy',ascending=False)
display(audi.iloc[0:5,:])
```

```python
# 5. mpg 데이터는 연비를 나타내는 변수가 2개입니다. 
#    두 변수를 각각 활용하는 대신 하나의 통합 연비 변수를 만들어 사용하려 합니다. 
#    평균 연비 변수는 두 연비(고속도로와 도시)의 평균을 이용합니다. 
#    회사별로 "suv" 자동차의 평균 연비를 구한후 내림차순으로 정렬한 후 1~5위까지 데이터를 출력하세요.

mpg['avg'] = (mpg['cty'] + mpg['hwy'])/2
suv = mpg.loc[mpg['clas'] == 'suv', ['avg']].groupby(mpg['manufacturer'])
suv = suv.mean().sort_values('avg', ascending=False)
display(suv.iloc[0:5,:])
```

```python
# 6. mpg 데이터의 class는 "suv", "compact" 등 자동차의 특징에 따라 
#    일곱 종류로 분류한 변수입니다. 어떤 차종의 도시 연비가 높은지 비교하려 합니다. 
#    class별 cty 평균을 구하고 cty 평균이 높은 순으로 정렬해 출력하세요.

clas = mpg.loc[:, ['cty']].groupby(mpg['clas'])
clas = clas.mean().sort_values('cty', ascending=False)
clas.columns = ['cty 평균']
display(clas)
```

```python
# 7. 어떤 회사 자동차의 hwy(고속도로 연비)가 가장 높은지 알아보려 합니다. 
#    hwy(고속도로 연비) 평균이 가장 높은 회사 세 곳을 출력하세요.

maf = mpg.loc[:, ['hwy']].groupby(mpg['manufacturer'])
maf = maf.mean().sort_values('hwy', ascending=False)
maf.columns = ['hwy 평균']
display(maf.iloc[:3,:])
```

```python
# 8. 어떤 회사에서 "compact" 차종을 가장 많이 생산하는지 알아보려고 합니다. 
#    각 회사별 "compact" 차종 수를 내림차순으로 정렬해 출력하세요.

compact = mpg.loc[mpg['clas'] == 'compact', ['avg']].groupby(mpg['manufacturer'])
compact = compact.count().sort_values(by='avg', ascending=False)
compact.columns = ['count']
display(compact)
```

# Pandas Exercise - 3

```python
import pandas as pd
import savReaderWriter

with savReaderWriter.SavReader('Koweps/Koweps_hpc10_2015_beta1.sav', ioUtf8 = True) as reader:
    df = pd.DataFrame(reader.all(), columns = [s for s in reader.header])
```

```python
# 1. 성별에 따른 월급 차이
#    과거에 비해 여성의 사회 진출이 활발하지만 직장에서의
#    위상에서는 여전히 차별이 존재하고 있는것이 사실.
#    실제로 그러한지 월급의 차이를 이용하여 사실을 확인해보자

man = df.loc[df['h10_g3'] == 1, ['p1002_8aq1']].mean()[0]
woman = df.loc[df['h10_g3'] == 2, ['p1002_8aq1']].mean()[0]

print("남성 평균 월급 : {:.5f}만원".format(man))
print("여성 평균 월급 : {:.5f}만원".format(woman))
print("월급 차이 : {:.5f}만원".format(man - woman))
```

```python
# 2. 나이와 월급의 관계
#    평균적으로 몇 살 때 월급을 가장 많이 받을까? 또 그때의 평균 월급은 얼마인가?

df['age'] = 2021 - df['h10_g4'] + 1
ageAvg = df.loc[:, ['p1002_8aq1']].groupby(df['age'])
ageAvg = ageAvg.mean().sort_values('p1002_8aq1', ascending=False)

print("월급을 가장 많이 받는 나이는 : {}세, 월급 : {:.5f}만원".format(ageAvg.index[0], ageAvg.iloc[0].values[0]))
```

```python
# 3. 연령대에 따른 월급 차이
#    30세 미만을 초년(young), 
#    30~59세 : 중년(middle), 
#    60세 이상 : 노년(old)
#    위의 범주로 연령대에 따른 월급의 차이를 알아보자

young = df.loc[df['age'] < 30, ['p1002_8aq1']].mean()[0]
middle = df.loc[(df['age'] >= 30) & (df['age'] <= 59), ['p1002_8aq1']].mean()[0]
old = df.loc[df['age'] >= 60, ['p1002_8aq1']].mean()[0]

print("초년(young) 평균 월급 : {:.5f}만원".format(young))
print("중년(middle) 평균 월급 : {:.5f}만원".format(middle))
print("노년(old) 평균 월급 : {:.5f}만원".format(old))
```

```python
# 4. 연령대 및 성별 월급 차이
#    성별 월급 차이는 연령대에 따라 다른 양상을 보일 수 있습니다.
#    성별 월급 차이가 연령대에 따라 다른지 분석해보자

#    기존에는 3그룹(초년,중년,노년)이었지만 이젠 6그룹으로
#    그룹핑을 해야 한다.(초년남성,초년여성,..)

youngM = df.loc[(df['age'] < 30) & (df['h10_g3'] == 1), ['p1002_8aq1']].mean()[0]
youngW = df.loc[(df['age'] < 30) & (df['h10_g3'] == 2), ['p1002_8aq1']].mean()[0]
middleM = df.loc[(df['age'] >= 30) & (df['age'] <= 59) & (df['h10_g3'] == 1), ['p1002_8aq1']].mean()[0]
middleW = df.loc[(df['age'] >= 30) & (df['age'] <= 59) & (df['h10_g3'] == 2), ['p1002_8aq1']].mean()[0]
oldM = df.loc[(df['age'] >= 60) & (df['h10_g3'] == 1), ['p1002_8aq1']].mean()[0]
oldW = df.loc[(df['age'] >= 60) & (df['h10_g3'] == 2), ['p1002_8aq1']].mean()[0]

print("초년 남성 평균 월급 : {:.5f}만원".format(youngM))
print("초년 여성 평균 월급 : {:.5f}만원".format(youngW))
print("중년 남성 평균 월급 : {:.5f}만원".format(middleM))
print("중년 여성 평균 월급 : {:.5f}만원".format(middleW))
print("노년 남성 평균 월급 : {:.5f}만원".format(oldM))
print("노년 여성 평균 월급 : {:.5f}만원".format(oldW))
```

```python
# 5. 직업별 월급 차이
#    어떤 직업이 월급을 가장 많이 받을까?
#    직업별 월급을 분석해 보자
#    직업코드는 제공된 Koweps_Codebook.xlsx을 이용하면 
#    편하게 코드값을 이용 할 수 있습니다.

jobCode = pd.read_csv('Koweps/jobCode.csv')
df2 = pd.merge(df,jobCode, left_on='h10_eco9', right_on='code_job')
df2 = df2.loc[:, ['p1002_8aq1','h10_eco9','job']].groupby('job')
jobAvg = df2.mean().sort_values('p1002_8aq1', ascending=False)
jobAvg.columns = ['평균 월급','직업코드']
display(jobAvg)
```

```python
# 6. 성별 직업 빈도
#    성별로 어떤 직업이 가장 많을까?

df3 = pd.merge(df,jobCode, left_on='h10_eco9', right_on='code_job')
jobM = df3.loc[df['h10_g3'] == 1, ['h10_eco9','job']].groupby('job')
jobM = jobM.count().sort_values('h10_eco9', ascending=False)
jobM.columns = ['Count']
print("남성 직업 빈도")
display(jobM)

jobW = df3.loc[df['h10_g3'] == 2, ['h10_eco9','job']].groupby('job')
jobW = jobW.count().sort_values('h10_eco9', ascending=False)
jobW.columns = ['Count']
print("여성 직업 빈도")
display(jobW)
```

```python
# 7. 종교 유무에 따른 이혼율
#    종교가 있는 사람들이 이혼을 덜 할까??
# 결혼을 해야 이혼을 하므로 비해당과 미혼자는 제외

pre = df.loc[(df['h10_g11'] == 1) & (df['h10_g10'] != 0) 
             & (df['h10_g10'] != 5), ['h10_g10','h10_g11']].groupby('h10_g10')
pre = pre.count()
pre.columns = ['Count']

entire = 0
divorce = 0
for idx, count in zip(pre.index, pre['Count']):
    if idx == 3:
        divorce = count
    entire += count

print("종교인의 이혼율 : {:.5f}%".format((divorce/entire)*100))


ab = df.loc[(df['h10_g11'] == 2) & (df['h10_g10'] != 0) 
             & (df['h10_g10'] != 5), ['h10_g10','h10_g11']].groupby('h10_g10')
ab = ab.count()
ab.columns = ['Count']

entire = 0
divorce = 0
for idx, count in zip(ab.index, ab['Count']):
    if idx == 3:
        divorce = count
    entire += count

print("무교인의 이혼율 : {:.5f}%".format((divorce/entire)*100))
```

```python
# 8. 지역별 연령대 비율
#    노년층이 많은 지역은 어디일까?
# young : 0, middle = 1, old = 2

df8 = df
df8['연령대'] = 0
df8.loc[df['age'] < 30, ['연령대']] = 0
df8.loc[(df['age'] >= 30) & (df['age'] <= 59), ['연령대']] = 1
df8.loc[df['age'] >= 60, ['연령대']] = 2

df8 = df8.loc[:, ['연령대']].groupby([df8['h10_reg7'], df8['연령대']])
df8 = df8.count()
df8.columns = ['Count']
display(df8)

maximum = 0
t = tuple()
for idx, count in zip(df8.index, df8['Count']):
    if count > maximum:
        maximum = count
        t = idx
# 1. 서울          2. 수도권(인천/경기)    3. 부산/경남/울산   4.대구/경북   
# 5. 대전/충남   6. 강원/충북               7.광주/전남/전북/제주도
print("노년층이 가장 많은 지역 : {}, {}명".format(t[0], maximum))
```
