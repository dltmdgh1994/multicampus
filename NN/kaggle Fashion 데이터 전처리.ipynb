```python
import pandas as pd

images = pd.read_csv('C:/Users/s_csmscox/jupyterSave/kFashion/kaggle_fashion/images.csv')
styles = pd.read_csv('C:/Users/s_csmscox/jupyterSave/kFashion/kaggle_fashion/styles.csv')
```

```python
data = styles.loc[(styles['subCategory'] == 'Topwear') |
                  (styles['subCategory'] == 'Bottomwear') |
                  (styles['subCategory'] == 'Shoes') |
                  (styles['subCategory'] == 'Bags') |
                  (styles['subCategory'] == 'Watches') |
                  (styles['subCategory'] == 'Innerwear') |
                  (styles['subCategory'] == 'Sandal') |
                  (styles['subCategory'] == 'Flip Flops') |
                  (styles['subCategory'] == 'Jewellery') |
                  (styles['subCategory'] == 'Belts') |
                  (styles['subCategory'] == 'Eyewear') |
                  (styles['subCategory'] == 'Wallets') |
                  (styles['subCategory'] == 'Socks') |
                  (styles['subCategory'] == 'Dress'),['id', 'subCategory']]
data
```

```python
data['subCategory'].loc[data['subCategory'] == 'Topwear'] = 0
data['subCategory'].loc[data['subCategory'] == 'Shoes'] = 1
data['subCategory'].loc[data['subCategory'] == 'Sandal'] = 1
data['subCategory'].loc[data['subCategory'] == 'Flip Flops'] = 1
data['subCategory'].loc[data['subCategory'] == 'Bags'] = 2
data['subCategory'].loc[data['subCategory'] == 'Bottomwear'] = 3
data['subCategory'].loc[data['subCategory'] == 'Watches'] = 4
data['subCategory'].loc[data['subCategory'] == 'Jewellery'] = 4
data['subCategory'].loc[data['subCategory'] == 'Belts'] = 4
data['subCategory'].loc[data['subCategory'] == 'Innerwear'] = 5
data['subCategory'].loc[data['subCategory'] == 'Eyewear'] = 6
data['subCategory'].loc[data['subCategory'] == 'Wallets'] = 7
data['subCategory'].loc[data['subCategory'] == 'Socks'] = 8
data['subCategory'].loc[data['subCategory'] == 'Dress'] = 9
```

```python
import shutil
from sklearn.model_selection import train_test_split

for i in range(0,10):
    sub = data.loc[data['subCategory']==i, :]

    train, test = train_test_split(sub, test_size=0.2, random_state=42)
    train, valid = train_test_split(train, test_size=0.2, random_state=42)

    dst = 'C:/Users/s_csmscox/jupyterSave/kFashion/kaggle_fashion/train/'

    for file_name, category_id in zip(train['id'],train['subCategory']):
        src = 'C:/Users/s_csmscox/jupyterSave/kFashion/kaggle_fashion/images/'+str(file_name)+'.jpg'
        try:
            shutil.copy(src, (dst + str(category_id ) + '/'))
        except:
            pass

    dst = 'C:/Users/s_csmscox/jupyterSave/kFashion/kaggle_fashion/test/'

    for file_name, category_id in zip(test['id'],test['subCategory']):
        src = 'C:/Users/s_csmscox/jupyterSave/kFashion/kaggle_fashion/images/'+str(file_name)+'.jpg'
        try:
            shutil.copy(src, (dst + str(category_id ) + '/'))
        except:
            pass

    dst = 'C:/Users/s_csmscox/jupyterSave/kFashion/kaggle_fashion/valid/'

    for file_name, category_id in zip(valid['id'],valid['subCategory']):
        src = 'C:/Users/s_csmscox/jupyterSave/kFashion/kaggle_fashion/images/'+str(file_name)+'.jpg'
        try:
            shutil.copy(src, (dst + str(category_id ) + '/'))
        except:
            pass
```

```python
import numpy as np

sub = styles['subCategory']
uni = sub.unique()
```

```python
cnt = []
product = []

for i in uni:
    cnt.append(0)
    product.append(i)

a = -1

for i in uni:
    a += 1
    for j in sub:
        if i == j:
            cnt[a] += 1
```

```python
df1 = pd.DataFrame(product, columns=['product'])
df2 = pd.DataFrame(cnt, columns=['cnt'])

df = pd.concat([df1, df2], axis=1)
df.sort_values(by=['cnt'], axis=0, inplace=True,  ascending=False)
df
```
