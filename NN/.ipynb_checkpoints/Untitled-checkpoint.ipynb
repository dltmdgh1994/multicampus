```python
import json

# images => file_name, id
# annotations => segmentation, category_id, image_id, id
# categories => id, name

with open('C:/Users/s_csmscox/jupyterSave/kFashion/dacon/data/train.json', 'r') as f:

    json_data = json.load(f)
```

```python
import pandas as pd
        
category_id = []
file_name = []
x_mins = []
y_mins = []
x_maxs = []
y_maxs = []

for i in json_data['images']:
    
    file_name.append(i['file_name'])
    img_id = i['id']
    c_id = []
    x_min = []
    y_min = []
    x_max = []
    y_max = []
    
    for j in json_data['annotations']:
        
        if img_id == j['image_id']:
            c_id.append(j['category_id'])
            bbox = j['bbox']
            
            # 바운딩 박스 좌표가 음수인 경우가 있어서 0으로 수정
            if bbox[0] < 0:
                bbox[0] = 1
            
            if bbox[1] < 0:
                bbox[1] = 1
            
            x_min.append(bbox[0])
            y_min.append(bbox[1])
            x_max.append(bbox[0]+bbox[2])
            y_max.append(bbox[1]+bbox[3])
        
        
    category_id.append(c_id)
    x_mins.append(x_min)
    y_mins.append(y_min)
    x_maxs.append(x_max)
    y_maxs.append(y_max)
        
y_mins
```

```python
# 이미지 크기가 800 * 800인 아닌 경우가 있어서 따로 빼줌

fname = []
height = []
width = []

no800 = dict()

for i in json_data['images']:
    
    if (i['height'] != 800) or (i['width'] != 800):
        fname.append(i['file_name'])
        height.append(i['height'])
        width.append(i['width'])
        
no800['file_name'] = fname
no800['height'] = height
no800['width'] = width

no800
```

```python
# 현재 경로 바꾸기

import os

os.chdir('C:/Users/s_csmscox/jupyterSave/kFashion/dacon/data/train_set/')
os.getcwd()
```

```python
from xml.etree.ElementTree import Element, SubElement, ElementTree

no800_len = len(no800['height'])

for f_name, c_ids, xmins, ymins, xmaxs, ymaxs  in zip(file_name, category_id, x_mins, y_mins, x_maxs, y_maxs):
    root = Element('annotation')
    SubElement(root, 'folder').text = 'train_set'
    SubElement(root, 'filename').text = f_name
    SubElement(root, 'path').text = './train_set/' +  f_name
    source = SubElement(root, 'source')
    SubElement(source, 'database').text = 'Unknown'

    size = SubElement(root, 'size')
    if f_name in no800['file_name']:
        for i in range(0, no800_len):
            if no800['file_name'][i] == f_name:
                SubElement(size, 'width').text = str(no800['width'][i])
                SubElement(size, 'height').text = str(no800['height'][i])
                break
    else:
        SubElement(size, 'width').text = str(800)
        SubElement(size, 'height').text = str(800)
    SubElement(size, 'depth').text = '1'

    SubElement(root, 'segmented').text = '0'

    for c_id, xmin, ymin, xmax, ymax in zip(c_ids, xmins, ymins, xmaxs, ymaxs):
        obj = SubElement(root, 'object')
        SubElement(obj, 'name').text = str(c_id)
        SubElement(obj, 'pose').text = 'Unspecified'
        SubElement(obj, 'truncated').text = '0'
        SubElement(obj, 'difficult').text = '0'
        bbox = SubElement(obj, 'bndbox')
        SubElement(bbox, 'xmin').text = str(xmin)
        SubElement(bbox, 'ymin').text = str(ymin)
        SubElement(bbox, 'xmax').text = str(xmax)
        SubElement(bbox, 'ymax').text = str(ymax)

    tree = ElementTree(root)
    tree.write('./' + f_name[:-4] +'.xml',encoding='utf-8')
```
